```python
custom_colours = ['#74a09e','#98e2c6','#f2a553','#c14953']
sns.set_theme(style="whitegrid",palette=custom_colours)
```

#### Loading the dataset

```python
data = []
with open('../paruvendu_api/outputs/data_harvest_full.json','r') as sample:
    for line in sample:
        data.append(json.loads(line.strip()))
df = pd.DataFrame.from_dict(data)
```

```python
df.drop(['status'], axis=1)
```

```python
url_list = df['url'].tolist()
brand_model_list = []
for url in url_list:
    url = url.split('/')
    brand_model = url[5] + "/" + url[6]
    brand_model_list.append(brand_model)
df['version'] = brand_model_list
df.head()
```
